```python
import tensorflow as tf
import numpy as np
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
# Generate sample data
num_samples = 1000
side1 = np.random.uniform(1, 10, num_samples)
side2 = np.random.uniform(1, 10, num_samples)
third_side = np.abs(side1 - side2) + np.random.uniform(0.001, 0.1, num_samples)

```

```python
# Prepare data
X = np.column_stack((side1, side2))
y = np.column_stack((third_side, side1 + side2, side1 + side2 - third_side))
```

```python
# Split data into train and test sets
X_train, X_test = X[:800], X[800:]
y_train, y_test = y[:800], y[800:]
```

```python
# Build the TensorFlow model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(2,)),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(32, activation='relu'),
    tf.keras.layers.Dense(3)  # 3 outputs for third side, perimeter, and area
])
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=500, batch_size=32, verbose=1)
```

```python
# Evaluate the model
loss = model.evaluate(X_test, y_test, verbose=0)
print(f"Test Loss: {loss:.4f}")
```

```python
# Predict using the trained model
predictions = model.predict(X_test)
```

```python
# Calculate Mean Squared Error (MSE) and Mean Absolute Error (MAE)
mse = mean_squared_error(y_test, predictions)
mae = mean_absolute_error(y_test, predictions)

print(f"Mean Squared Error (MSE): {mse:.4f}")
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```

```python
# Print some example predictions
for i in range(5):
    print(f"Sample {i+1}:")
    print(f"Predicted Third Side: {predictions[i][0]:.2f} (Actual: {y_test[i][0]:.2f})")
    print(f"Predicted Perimeter: {predictions[i][1]:.2f} (Actual: {y_test[i][1]:.2f})")
    print(f"Predicted Area: {predictions[i][2]:.2f} (Actual: {y_test[i][2]:.2f})")
    print()
```
